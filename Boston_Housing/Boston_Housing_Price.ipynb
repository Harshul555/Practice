```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

__Boston House Price Dataset__

```python
bn = pd.read_csv("Boston.csv")
```

```python
bn.drop(columns= ["Unnamed: 0"], axis=0, inplace = True)
bn.head()
```

```python
# statistical info
bn.describe()
```

```python
# datatype info
bn.info()
```

```python
# check for null values
bn.isnull().sum()
```

```python
# create box plots 
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize =(20,10))
index = 0
ax = ax.flatten()

for col, value in bn.items():
    sns.boxplot(y = col, data = bn, ax = ax[index])
    index += 1
plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)
```

```python
# create dist plot
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize =(20,10))
index = 0
ax = ax.flatten()

for col, value in bn.items():
    if col == "chas":
        continue
    sns.distplot(value, ax = ax[index])
    plt.title(col)
    index += 1
plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)
```

```python
# Min-max Normalization

cols = ['crim', 'zn', 'tax', 'black']
for col in cols:
    # find minimum and maximum of that column
    minimum = min(bn[col])
    maximum = max(bn[col])
    bn[col] = (bn[col] - minimum) / (maximum - minimum)
```

```python
# create dist plot after normalization
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize =(20,10))
index = 0
ax = ax.flatten()

for col, value in bn.items():
    if col == "chas":
        continue
    sns.distplot(value, ax = ax[index])
    plt.title(col)
    index += 1
plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)
```

```python
# standardization 
from sklearn import preprocessing as prps
scalar = prps.StandardScaler()

# fit our data
scaled_cols = scalar.fit_transform(bn[cols])
scaled_cols = pd.DataFrame(scaled_cols, columns = cols)
scaled_cols.head()
```

```python
for col in cols:
    bn[col] = scaled_cols[col]
```

```python
# create dist plot after normalization and standardization
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize =(20,10))
index = 0
ax = ax.flatten()

for col, value in bn.items():
    if col == "chas":
        continue
    sns.distplot(value, ax = ax[index])
    plt.title(col)
    index += 1
plt.tight_layout(pad = 0.5, w_pad = 0.7, h_pad = 5.0)
```

__Input Split__

```python
X = bn.drop(columns = ['medv', 'rad'], axis = 1)
y = bn['medv']
```

__Model Training__

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
def train(model, X, y):
    # train the model
    x_train, x_test, y_train, y_test = train_test_split(X, y, random_state = 42)
    model.fit(x_train,y_train)
    
    # predict the training data
    pred = model.predict(x_test)
    
    # perform cross-validation
    cv_score = cross_val_score(model, X,y, scoring = "neg_mean_squared_error", cv = 5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE: ", mean_squared_error(y_test, pred))
    print("CV Score", cv_score)
```

```python
from sklearn.linear_model import LinearRegression as LR
model = LR(normalize = True)
train(model, X, y)
coef = pd.Series(model.coef_, X.columns).sort_values()
coef.plot(kind = "bar", title = "Model Coefficients")
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")
```

```python
import xgboost as xgb
model = xgb.XGBRegressor()
train(model, X, y)
coef = pd.Series(model.feature_importances_, X.columns).sort_values(ascending = False)
coef.plot(kind = "bar", title = "Feature Importance")
```
